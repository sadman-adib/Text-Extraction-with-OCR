```python
# 🚀 Step 3: Import Libraries & Load Models
import pytesseract
from PIL import Image
from pdf2image import convert_from_path
from transformers import pipeline
from google.colab import files

# Define Model Name
MODEL_NAME = "meta-llama/Llama-3.2-1B"  # Change if needed

# Load LLaMA 3.2 Using Pipeline
def load_llama_pipeline():
    print("🔄 Loading LLaMA 3.2 Pipeline...")
    text_generator = pipeline("text-generation", model=MODEL_NAME, token=HF_TOKEN, device=0)
    return text_generator

llama_pipeline = load_llama_pipeline()
```

```python
# 🚀 Step 4: Upload File
def upload_file():
    print("📂 Please upload an image or PDF file...")
    uploaded = files.upload()
    return list(uploaded.keys())[0]

file_name = upload_file()
```

```python
# 🚀 Step 5: Extract Text from Image or PDF
def extract_text(file_path):
    if file_path.lower().endswith(".pdf"):
        print("📄 Extracting text from PDF...")
        images = convert_from_path(file_path)
        extracted_text = "\n\n".join([pytesseract.image_to_string(img) for img in images])
    else:
        print("🖼️ Extracting text from Image...")
        extracted_text = pytesseract.image_to_string(Image.open(file_path))

    return extracted_text.strip()

extracted_text = extract_text(file_name)
```

```python
# 🚀 Step 6: Process Extracted Text Using LLaMA 3.2 Pipeline
def process_text_with_llama(text):
    print("🤖 Processing text with LLaMA 3.2 pipeline...")
    output = llama_pipeline(text, max_length=500, do_sample=True)[0]["generated_text"]
    return output

processed_text = process_text_with_llama(extracted_text)
```

```python
# 🚀 Step 7: Display Results
print("\n📝 Extracted Text:\n", extracted_text)
print("\n🤖 Processed Text with LLaMA:\n", processed_text)
```
